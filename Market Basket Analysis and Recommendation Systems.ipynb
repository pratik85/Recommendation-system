```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('GroceryStoreDataSet.csv',delimiter=',',names=['products'],header=None)
```

```python
df.head()
```

```python
df.values
```

```python
data=list(df['products'].apply(lambda x: x.split(',')))
data
          
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
te_data=te.fit(data).transform(data)
df1=pd.DataFrame(te_data,columns=te.columns_)
df1
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df2 = apriori(df1,min_support=0.01,use_colnames=True)
df2
```

```python
DF3 = association_rules(df2, metric='lift')
DF3.head(10)
```

```python
DF3[ (DF3['lift'] > 1.1) & (DF3['confidence'] > 0.4)]
```
